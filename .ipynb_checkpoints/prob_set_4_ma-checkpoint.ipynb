### Exercise. 4

```python
def ma2sim(theta=[1, -2, 1.35], mean=0.7):
    eps = np.random.randn(102)
    series = []
    for i in range(2, 102):
        y = 0.7 + theta[0] * eps[i] + theta[1] * eps[i-1] + theta[2] * eps[i-2]
        series.append(y)
    sm.tsa.graphics.plot_acf(series, lags=10, zero=True)
    plt.show()
    sm.tsa.graphics.plot_pacf(series, lags=10, zero=True, method="ywmle")
    plt.show()
    return series
```

```python
sim1 = ma2sim()
```
